# Stage 2 : Tokenization

```python
clean_text_2= []

from nltk.tokenize import sent_tokenize,word_tokenize

import nltk
nltk.download('punkt')
```

```python
clean_text_2= []

from nltk.tokenize import sent_tokenize,word_tokenize

import nltk
nltk.download('punkt_tab') # Download the required resource

sent_tok=[]
for sent in clean_text_1:
  sent=sent_tokenize(sent)
  sent_tok.append(sent)
```

```python
sent_tok
```

```python
# word tokenize

clean_text_2=[word_tokenize(i) for i in clean_text_1]
```

```python
clean_text_2
```

```python
import re

clean_text_3=[]
for words in clean_text_2:
  clean=[]
  for w in words:
    res=re.sub(r'[^\w\s]','',w)
    if res != '':
      clean.append(res)
  clean_text_3.append(clean)
```

```python
clean_text_3
```

# Satge 4 : Stopwords Removal

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
clean_text_4=[]

for words in clean_text_3:
  w=[]
  for word in words:
    if word not in stopwords.words('english'):
      w.append(word)
  clean_text_4.append(w)
```

```python
clean_text_4
```
